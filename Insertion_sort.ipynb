```python
import timeit
import matplotlib.pyplot as plt

# Insertion Sort function
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j = j - 1
        arr[j + 1] = key

# Get user input for the number of elements in the array
n = int(input("Enter number of elements: "))

# Get user input for the elements of the array
arr = []
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    arr.append(element)

# Apply Insertion Sort
insertionSort(arr)

# Print sorted array
print("Sorted array is:")
for i in range(len(arr)):
    print("%d" % arr[i])

# Measuring time complexity with different array sizes for plotting
list_lengths = [100, 200, 300, 400, 500]  # Example array sizes to test
time_taken = []

for length in list_lengths:
    arr = [random.randint(1, 1000) for _ in range(length)]  # Generate a random array of 'length' elements

    # Measure time taken to sort the array using timeit
    setup_code = "from __main__ import insertionSort"
    test_code = f"insertionSort({arr})"
    execution_time = timeit.timeit(test_code, setup=setup_code, number=1)
    time_taken.append(execution_time)

# Plotting the time complexity graph
plt.plot(list_lengths, time_taken, marker='o')
plt.xlabel('Size of array')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of Insertion Sort')
plt.show()
```

```python
import timeit
import random
import matplotlib.pyplot as plt

# Heapify function to maintain the heap property
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    if left < n and arr[i] < arr[left]:
        largest = left
    if right < n and arr[largest] < arr[right]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

# Heap Sort function
def heapSort(arr):
    n = len(arr)
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

# Get user input for array sizes
list_lengths_input = input("Enter array sizes (comma separated, e.g., 10000, 20000, 30000): ")
list_lengths = list(map(int, list_lengths_input.split(',')))

# Get user input for the range of numbers
min_value = int(input("Enter the minimum value of the random numbers: "))
max_value = int(input("Enter the maximum value of the random numbers: "))

time_taken = []

# Loop through each list length and perform heap sort
for length in list_lengths:
    arr = [random.randint(min_value, max_value) for _ in range(length)]

    # Time the sorting operation using timeit
    setup_code = "from __main__ import heapSort"
    test_code = f"heapSort({arr})"
    execution_time = timeit.timeit(test_code, setup=setup_code, number=1)
    time_taken.append(execution_time)

    # Print the sorted array for each list (optional, can slow down execution for large lists)
    print(f"Sorted array of size {length}: {arr[:20]}...")  # Print only first 20 elements for large arrays

# Plotting the time taken vs array size graph
plt.plot(list_lengths, time_taken, marker='o')
plt.xlabel('Size of array')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of Heap Sort')
plt.show()
```
