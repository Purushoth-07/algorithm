```python
import matplotlib.pyplot as plt
def linear_search(arr,x):
  for i in range(len(arr)):
    if arr[i] == x:
      return i
  return -1
arr=[2,4,6,8,45]
x=652
result=linear_search(arr,x)
if result == -1:
    print("element not found")
else:
    print("element found at index",result)
n=[10,20,50,100,200,300,450,500]
time=[0.0001,0.0003,0.0008,0.0015,0.0030,0.0035,0.0050,0.0060]
plt.plot(n,time)
plt.xlabel("n")
plt.ylabel("time")
plt.title("time Complexity linear search")
plt.show()
```
